```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.2)
```

```python
df = pd.read_csv('50ulke.csv')
df.head()
```

# distplot()

```python
sns.distplot(df['Growth'])
```

```python
sns.distplot(df['National Income'])
```

```python
sns.distplot(df['Per person'])
```

# kdeplot()

```python
sns.kdeplot(df['National Income'], shade=True)
```

```python
sns.kdeplot(df['Per person'], shade=True)
```

```python
sns.kdeplot(df['Growth'], shade=True)
```

# jointplot()

```python
sns.jointplot(x='Growth', y='National Income', data=df)
```

```python
sns.jointplot(x='Growth', y='Per person', data=df, kind='kde')
```

```python
sns.jointplot(x='National Income', y='Per person', data=df, kind='hex')
```

```python
sns.pairplot(df)
```
