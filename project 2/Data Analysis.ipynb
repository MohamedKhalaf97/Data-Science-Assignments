# Grouped Summuries

```python
df.groupby(['Warehouse_block','Mode_of_Shipment','Product_importance'])['Discount_offered'].agg([min, max, np.mean, np.median,sum, np.std, np.var])
```

```python
df.groupby(['Warehouse_block','Mode_of_Shipment','Product_importance'])['Discount_offered'].describe()
```

```python
df.groupby(['Warehouse_block','Mode_of_Shipment','Product_importance'])['Discount_offered'].describe().transpose()
```

```python
df.groupby(['Warehouse_block','Mode_of_Shipment','Product_importance'])['Discount_offered'].describe().transpose()['B']
```

```python
regionFilter = df.groupby('Warehouse_block')['Discount_offered'].max()
regionFilter
```

# Pivot Table

```python
df.pivot_table(index=["Warehouse_block",'Mode_of_Shipment','Product_importance'], aggfunc=[min, max, np.mean, np.median,sum, np.std, np.var], values="Discount_offered")
```

```python
pd.pivot_table(df, index=['Mode_of_Shipment','Product_importance'], columns='Warehouse_block', values='Discount_offered')
```
